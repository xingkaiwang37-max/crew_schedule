```python
from datetime import datetime
flight_df = pd.read_csv("D:\Workspace\CTS\CTS20250606ç«žèµ›æ•°æ®\\flight.csv")
flight_dict = {}
for index, row in flight_df.iterrows():
    std_dt = datetime.strptime(row['std'], "%Y/%m/%d %H:%M")
    # å°†datetimeå¯¹è±¡è½¬æ¢ä¸ºæ—¶é—´æˆ³ï¼ˆç§’çº§ï¼‰
    std = std_dt.timestamp()

    sta_dt = datetime.strptime(row['sta'], "%Y/%m/%d %H:%M")
    # å°†datetimeå¯¹è±¡è½¬æ¢ä¸ºæ—¶é—´æˆ³ï¼ˆç§’çº§ï¼‰
    sta = sta_dt.timestamp()
    flight_dict[row['id']] = {"depaAirport": row['depaAirport'], "arriAirport": row['arriAirport'], "std": int(std*1000), "sta": int(sta*1000)}
flight_dict
```

```python
businfo_df = pd.read_csv("D:\Workspace\CTS\CTS20250606ç«žèµ›æ•°æ®\\busInfo.csv")
businfo_dict = {}
for index, row in businfo_df.iterrows():
    td_dt = datetime.strptime(row['td'], "%Y/%m/%d %H:%M")
    # å°†datetimeå¯¹è±¡è½¬æ¢ä¸ºæ—¶é—´æˆ³ï¼ˆç§’çº§ï¼‰
    depaTime = td_dt.timestamp()
    ta_dt = datetime.strptime(row['ta'], "%Y/%m/%d %H:%M")
    # å°†datetimeå¯¹è±¡è½¬æ¢ä¸ºæ—¶é—´æˆ³ï¼ˆç§’çº§ï¼‰
    arriTime = ta_dt.timestamp()
    businfo_dict[row['id']] = {"depaAirport": row['depaAirport'], "arriAirport": row['arriAirport'], "depaTime": int(depaTime*1000), "arriTime": int(arriTime*1000)}
businfo_dict
```

```python
yAxisCategories = []
yAxisCategories_dict = {}
for index, row in df.iterrows():
    if row['crewId'] not in yAxisCategories_dict:
        yAxisCategories_dict[row['crewId']] = len(yAxisCategories)
        yAxisCategories.append(row['crewId'])
```

```python
seriesData = []
for index, row in df.iterrows():
    start = None
    end = None
    color = None
    # if row['taskType'] == 'Flt':
    if row['isDDH'] == 0:
        start = flight_dict[row['taskId']]['std']
        end = flight_dict[row['taskId']]['sta']
        dep = flight_dict[row['taskId']]['depaAirport']
        arr = flight_dict[row['taskId']]['arriAirport']
        color = '#7cb5ec'
    elif row['isDDH'] == 1:
        if row['taskType'] == 'Flt':
            start = flight_dict[row['taskId']]['std']
            end = flight_dict[row['taskId']]['sta']
            dep = flight_dict[row['taskId']]['depaAirport']
            arr = flight_dict[row['taskId']]['arriAirport']
            color = '#f15b4f'
        else:
            start = businfo_dict[row['taskId']]['depaTime']
            end = businfo_dict[row['taskId']]['arriTime']
            dep = businfo_dict[row['taskId']]['depaAirport']
            arr = businfo_dict[row['taskId']]['arriAirport']
            color = '#90ed7d'
    seriesData.append({
        "start": int(start),
        "end": int(end),
        "y": yAxisCategories_dict[row['crewId']],
        "color": color,
        "taskType": row['taskType'],
        "taskId": row['taskId'],
        "dep": dep,
        "arr": arr,
    })
result = {"yAxisCategories": yAxisCategories, "seriesData": seriesData}
result
```

```python
with open('multi_base_gantt_chart_data.json', 'w') as f:
    import json
    json.dump(result, f)
```
