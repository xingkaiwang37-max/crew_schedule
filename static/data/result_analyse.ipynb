```python
import pandas as pd
flight_df = pd.read_csv("flight.csv")
flight_data = {}
for _, row in flight_df.iterrows():
    flight_data[row['id']] = {
        'std': row['std'],
        'sta': row['sta'],
        'depaAirport': row['depaAirport'],
        'arriAirport': row['arriAirport'], 
        'flyTime': row['flyTime']
    }
flight_data
```

```python
import pandas as pd
bus_info_df = pd.read_csv("busInfo.csv")
bus_info_df
```

```python
import pandas as pd
crew_df = pd.read_csv("crew.csv")
crew_df
```

```python
result = pd.read_csv('multi_base/rosterResult.csv')
result['taskType'] = result['taskId'].apply(lambda x: x.split('_')[0])
result = pd.merge(result, flight_df, left_on='taskId', right_on='id', how='left')
result = pd.merge(result, bus_info_df, left_on='taskId', right_on='id', how='left')
result = pd.merge(result, crew_df, left_on='crewId', right_on='crewId', how='left')
# result['std_day'] = result['std'].apply(lambda x: int(x.split(' ')[0].split('-')[-1]))
result['sta'] = pd.to_datetime(result['sta'])
result['std'] = pd.to_datetime(result['std'])
result['ta'] = pd.to_datetime(result['ta'])
result['td'] = pd.to_datetime(result['td'])
result['sta'] = result['sta'].fillna(result['ta'])
result['std'] = result['std'].fillna(result['td'])

result['depaAirport'] = result['depaAirport_x'].fillna(result['depaAirport_y'])
result['arriAirport'] = result['arriAirport_x'].fillna(result['arriAirport_y'])
# result = result.sort_values(by='sta')
result
```
